```python
from astropy import units as u   #Conversion y desarrollo de operaciones aritmÃ©ticas de instancias Quantity
from astropy.table import Table
import numpy as np
from matplotlib import pyplot as plt
import math as mt
#import sys
```

```python

resultList = Table.read("resultList.tbl", format = 'ascii')
OIII5007_WT = 5008.24
Hb_WT = 4862.683
NII6583_WT = 6585.28
Ha_WT = 6564.61
SII6717_WT = 6718.29
SII6731_WT = 6732.67
OIII4959_WT = 4960.295
OII3727_WT = 3727.092
OII3729_WT = 3729.875


%matplotlib

#fig = plt.figure()
#ax = fig.add_subplot(2,1,1)
#ax.set_yscale('log')

plt.title('BPT diagram 3')
plt.ylabel(r'Log([O III]/ H$\beta$)')

#plt.yscale('lin')
#plt.xscale('lin')

cooo = -1
Sy1 = []
Sy2 = []
Stb = []
HIIR = []
LINER=[]
OSy1 = []
OSy2 = []
OStb = []
OHIIR = []
OLINER=[]

#typeBPT = 0  #  NII 
# 1  #  SII OIII5007
# 2 O III5007+4959 and SII
#  3 OII 3727+3729
typeBPT = 3

limDis = 2.0

if typeBPT == 0:
    plt.xlabel(r'Log([N II]/H$\alpha$)')
elif (typeBPT == 1) or (typeBPT == 2):
    plt.xlabel(r'Log([S II]/H$\alpha$)')
elif typeBPT == 3:
    plt.xlabel(r'Log([O II]/H$\beta$)')

for iinn in range(len(resultList)):  #  objID in resultList:
    nameTable = "Tabla"+str(resultList['ObjID'][iinn])+".tbl"
#    nameTable = "Tabla"+str(objID[0])+".tbl"
    tableObjID = Table.read(nameTable, format = 'ascii')
    OII3727 = tableObjID["Flux"][0]
    OII3729 = tableObjID["Flux"][1]
    OIII5007 = tableObjID["Flux"][7]
    OIII4959 = tableObjID["Flux"][6]
    Hb = tableObjID["Flux"][4]
    NII6583 = tableObjID["Flux"][12]
    Ha = tableObjID["Flux"][11] 
    SII6717 = tableObjID["Flux"][13]
    SII6731 = tableObjID["Flux"][14]               # +++++++++++++++
    OII3727_W = tableObjID["Wavelength"][0]
    OII3729_W = tableObjID["Wavelength"][1]
    OIII5007_W = tableObjID["Wavelength"][7]
    OIII4959_W = tableObjID["Wavelength"][6]
    Hb_W = tableObjID["Wavelength"][4]
    NII6583_W = tableObjID["Wavelength"][12]
    Ha_W = tableObjID["Wavelength"][11]
    SII6717_W = tableObjID["Wavelength"][13]
    SII6731_W = tableObjID["Wavelength"][14]
    if typeBPT == 0:
        if OIII5007>0.0 and Hb>0.0 and NII6583>0.0 and Ha>0.0 and mt.fabs(OIII5007_WT-OIII5007_W)<limDis and mt.fabs(Hb_WT-Hb_W)<limDis and mt.fabs(NII6583_WT-NII6583_W)<limDis and mt.fabs(Ha_WT-Ha_W)<limDis:
            cooo = cooo + 1
            if resultList['type'][iinn] == 'NL SY1':
                Sy1.append(NII6583/Ha)
                OSy1.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'SY2':
                Sy2.append(NII6583/Ha)
                OSy2.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Starburst':
                Stb.append(NII6583/Ha)
                OStb.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'H II':
                HIIR.append(NII6583/Ha)
                OHIIR.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Liner':
                LINER.append(NII6583/Ha)
                OLINER.append(OIII5007/Hb)
    elif typeBPT == 1:
        if OIII5007>0.0 and Hb>0.0 and SII6717>0.0 and SII6731>0.0 and Ha>0.0 and mt.fabs(OIII5007_WT-OIII5007_W)<limDis and mt.fabs(Hb_WT-Hb_W)<limDis and mt.fabs(SII6731_WT-SII6731_W)<limDis and mt.fabs(SII6717_WT-SII6717_W)<limDis and mt.fabs(Ha_WT-Ha_W)<limDis:
    #    #    print NII6583/Ha, OIII5007/Hb, nameTable
            cooo = cooo + 1
            if resultList['type'][iinn] == 'NL SY1':
                Sy1.append((SII6717+SII6731)/Ha)
                OSy1.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'SY2':
                Sy2.append((SII6717+SII6731)/Ha)
                OSy2.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Starburst':
                Stb.append((SII6717+SII6731)/Ha)
                OStb.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'H II':
                HIIR.append((SII6717+SII6731)/Ha)
                OHIIR.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Liner':
                LINER.append((SII6717+SII6731)/Ha)
                OLINER.append(OIII5007/Hb)
    elif typeBPT == 2:
        if OIII4959>0.0 and OIII5007>0.0 and Hb>0.0 and SII6717>0.0 and SII6731>0.0 and Ha>0.0 and mt.fabs(OIII4959_WT-OIII4959_W)<limDis and mt.fabs(OIII5007_WT-OIII5007_W)<limDis and mt.fabs(Hb_WT-Hb_W)<limDis and mt.fabs(SII6731_WT-SII6731_W)<limDis and mt.fabs(SII6717_WT-SII6717_W)<limDis and mt.fabs(Ha_WT-Ha_W)<limDis:
    #    #    print NII6583/Ha, OIII5007/Hb, nameTable
            cooo = cooo + 1
            if resultList['type'][iinn] == 'NL SY1':
                Sy1.append((SII6717+SII6731)/Ha)
                OSy1.append((OIII5007+OIII4959)/Hb)
            if resultList['type'][iinn] == 'SY2':
                Sy2.append((SII6717+SII6731)/Ha)
                OSy2.append((OIII5007+OIII4959)/Hb)
            if resultList['type'][iinn] == 'Starburst':
                Stb.append((SII6717+SII6731)/Ha)
                OStb.append((OIII5007+OIII4959)/Hb)
            if resultList['type'][iinn] == 'H II':
                HIIR.append((SII6717+SII6731)/Ha)
                OHIIR.append((OIII5007+OIII4959)/Hb)
            if resultList['type'][iinn] == 'Liner':
                LINER.append((SII6717+SII6731)/Ha)
                OLINER.append((OIII5007+OIII4959)/Hb)
    elif typeBPT == 3:
        if OIII5007>0.0 and Hb>0.0 and OII3727>0.0 and OII3729>0.0 and mt.fabs(OIII5007_WT-OIII5007_W)<limDis and mt.fabs(Hb_WT-Hb_W)<limDis and mt.fabs(OII3727_WT-OII3727_W)<limDis and mt.fabs(OII3729_WT-OII3729_W)<limDis:
    #    #    print NII6583/Ha, OIII5007/Hb, nameTable
            cooo = cooo + 1
            if resultList['type'][iinn] == 'NL SY1':
                Sy1.append((OII3727+OII3729)/Hb)
                OSy1.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'SY2':
                Sy2.append((OII3727+OII3729)/Hb)
                OSy2.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Starburst':
                Stb.append((OII3727+OII3729)/Hb)
                OStb.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'H II':
                HIIR.append((OII3727+OII3729)/Hb)
                OHIIR.append(OIII5007/Hb)
            if resultList['type'][iinn] == 'Liner':
                LINER.append((OII3727+OII3729)/Hb)
                OLINER.append(OIII5007/Hb)

#s1 = plt.scatter(Sy1, OSy1, c='k', marker='o',label='NL SY1')
#s2 = plt.scatter(Sy2, OSy2, c='yellow', marker='D',label='SY2')
#st = plt.scatter(Stb, OStb, c='cyan', marker='v',label='Starburst')
#hii = plt.scatter(HIIR, OHIIR, c='blue', marker='s',label='H II')
#liner = plt.scatter(LINER, OLINER, c='w', marker='o',label='LINER')
s1 = plt.scatter(np.log10(Sy1), np.log10(OSy1), c='k', marker='o',label='NL SY1')
s2 = plt.scatter(np.log10(Sy2), np.log10(OSy2), c='yellow', marker='D',label='SY2')
st = plt.scatter(np.log10(Stb), np.log10(OStb), c='cyan', marker='v',label='Starburst')
hii = plt.scatter(np.log10(HIIR), np.log10(OHIIR), c='blue', marker='s',label='H II')
liner = plt.scatter(np.log10(LINER), np.log10(OLINER), c='w', marker='o',label='LINER')
x = np.log10(np.linspace(0.01, 1, 20))
if typeBPT == 0:
    y = 0.61/(x-0.05) + 1.30
    y1 = 0.61/(x-0.47) + 1.19
    plt.plot(x, y, c='r', label='SF galaxies')
    plt.plot(x, y1, c='dimgrey', ls='--', label='Composite galaxies')
    plt.text(-0.584, 1.72, 'AGN')
    plt.text(-1.12, -0.358, 'H II')
    plt.text(-0.2306, -0.55, 'Comp')
elif (typeBPT == 1) or (typeBPT == 2):
    y = 0.72/(x-0.32) + 1.30
    plt.plot(x, y, c='r', label='SF galaxies')
    plt.axvline(x=-0.4,c='dimgrey',lw=1.2,ls='--')
    x1 = np.linspace(-0.57, 0.5, 20)
    plt.plot(x1, [0.5]*len(x1), c='dimgrey',lw=1.2,ls='--')
    plt.text(0.0557, -0.106, 'LINER')
    plt.text(-1.125, 1.467, 'Seyfert')
    plt.text(-1.125, -0.734, 'H II')
elif typeBPT == 3:
    x = np.log10(np.linspace(0.01, 8, 20))
    y = 0.11/(x-0.92) + 0.85
    plt.plot(x, y, c='r', label='SF galaxies')
    plt.text(-0.872, -0.494, 'Star Forming')
    plt.text(-0.872, 1.1987, 'Seyfert 2')


plt.legend((s1, s2, st, hii, liner), ('NL SY1', 'SY2', 'Starburst', 'H II', 'LINER'),ncol=1, markerscale= 2, loc='upper right', fontsize=10)
print "Sample = ", str(len(resultList))
print "Number of data selected = ", cooo
print "NL SY1: ",str(len(Sy1))
print "SY2: ",str(len(Sy2))
print "Stb: ",str(len(Stb))
print "H II: ",str(len(HIIR))
print "LINER: ",str(len(LINER))
```
